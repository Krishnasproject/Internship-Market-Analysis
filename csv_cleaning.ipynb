```python
import pandas as pd

# Loading file
df = pd.read_csv('final_internship_data.csv')

# This part cleans the 'job description' column so it doesn't break Tableau
# It finds any hidden "Enter" or "New Line" keys and replaces them with a space
str_cols = df.select_dtypes(include=['object']).columns
for col in str_cols:
    df[col] = df[col].str.replace(r'\r\n|\r|\n', ' ', regex=True)

# Save the clean version
df.to_csv('cleaned_internship_data.csv', index=False)
print("Success! Your new file 'cleaned_internship_data.csv' is ready for Tableau.")
```

```python
import pandas as pd

# Load the file
df = pd.read_csv('cleaned_internship_data.csv')

# 1. Remove exact duplicates based on the Job URL
df = df.drop_duplicates(subset=['job_url'], keep='first')

# 2. Remove rows where the job_title or job_url is missing
df = df.dropna(subset=['job_url', 'job_title'])

# 3. Unify "Remote" and "Work from home"
df['location'] = df['location'].replace('Remote', 'Work from home')

# 4. Fill missing duration with 'Not Specified'
df['duration_text'] = df['duration_text'].fillna('Not Specified')

# Save the repaired file
df.to_csv('cleaned_internship_data.csv', index=False)

print(f"✅ Repair Complete!")
print(f"Original Rows: 2333")
print(f"Clean Unique Rows: {len(df)}")
print("Now go to Tableau and click REFRESH.")
```
