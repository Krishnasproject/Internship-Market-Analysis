```python
import pandas as pd

# Loading file
df = pd.read_csv('final_internship_data.csv')

# This part cleans the 'job description' column so it doesn't break Tableau
# It finds any hidden "Enter" or "New Line" keys and replaces them with a space
str_cols = df.select_dtypes(include=['object']).columns
for col in str_cols:
    df[col] = df[col].str.replace(r'\r\n|\r|\n', ' ', regex=True)

# Save the clean version
df.to_csv('cleaned_internship_data.csv', index=False)
print("Success! Your new file 'cleaned_internship_data.csv' is ready for Tableau.")
```
